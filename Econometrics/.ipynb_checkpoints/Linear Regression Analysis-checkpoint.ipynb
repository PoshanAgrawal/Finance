# Linear Regression

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf

plt.rcParams["figure.figsize"]=(12,9)
```

**Variables in the dataset** - *All variables are quaterly ratios*
1. Stock's excess return (%), (actual return - risk_free interest rate)
2. Sales growth (%),
3. DOW 30 Index Excess Retuen (%)
4. Pepsi's stock excess return (%)

Key assumption of linear regreession is that the endogenous variable has a linear relationship with the exogenous variable. 

**Check -** Scatterplot.<br>
If horizontal or vertical - no relationship - drop the variable.<br>
Finding Extreme Points:
1. Outlier: Point that doesn't follows the general trend.
2. Leverage Point: Extreme Value for exogenous variable.

Both may or maynot impact the regression result.

```python
#Importing Data
df= pd.read_excel("Data/Finance_coke.xlsx")
#KO_Q_EX_R: KO Quarterly Stock Excess Return
#Dow_Q_EX_R: DOW 30 Index Quarterly Excess Return
#Pepsi_Q_EX_R: Pepsi Quarterly Stock Excess Return
#REV_Q_R: KO Quarterly Revenue change % 
```

```python
df.head()
```

## Checking the Relationship

```python
#Create Scatter Plot:
scatt= df.plot.scatter(
    x="Coke_Q_EX_R",
    y= "Dow_Q_EX_R",
    title= "Scatterplot Coke Stock Excess Return vs Dow 30 Indedx Excess Return",
)
scatt.set(xlabel= "Cock Stock Excess Return", ylabel= "DOW 30 Index Excess Return")
plt.show()
```

```python
# **kwargs- Keyword arguments ae syntax that allows to pass dictionary of named argument to a function
def corr(x,y,**kwargs):
    #Calculate value - picks the value from the correlatin matrix
    coef= np.corrcoef(x,y)[0][1]
    
    #Make Label 
    # r" - (raw string literal) syntax is used to create strings where backslashes \ are treated as literal characters, rather than as escape characters. 
    label= r"$\rho$= " + str(round(coef,2))
    
    #Label to plot
    # gca- get current axes
    ax= plt.gca()
    ax.annotate(label, xy=(.3,.15), size = 20, xycoords= ax.transAxes)
```

```python
grid= sns.pairplot(
    df,vars=["Coke_Q_EX_R", "Dow_Q_EX_R", "Pepsi_Q_EX_R","REV_Q_R"], height =3
)

# Mapping the scatterplot and the pearson correlation coefficient to the uper triangle
grid= grid.map_upper(plt.scatter)
grid=grid.map_upper(corr)

# Mapping the histogram to the diagonal
grid=grid.map_diag(plt.hist)

# Mapping the density plot to the lower triangle
grid=grid.map_lower(sns.kdeplot) #kernel density estimation
```

## Regression Model

**KO Excess return = $b_{0}$ + $b_{1}$ Dow excess return + $b_{2}$ pepsi excess return +$b_{3}$ KO revenue growth**

1. KO excess return has a good correlations with Dow excess return anad pepsi excess return.
2. Ko excess return has vry poor correlation with KO revenue growth.

```python
#Linear Regression
result= smf.ols(
    "Coke_Q_EX_R ~ Dow_Q_EX_R + Pepsi_Q_EX_R + REV_Q_R", data=df
).fit()
result.summary()
```

```python
# Parameters with 6 significant digits
result.summary2().tables[1]
```

Based on the above table it can be seen that Dow excess return is only significant ( value of P>[t] < 0.05).

Before droping the other two variables things that are needed to be considered:

> 1. Variable transformation,


> 2. Correlation between the exogenous variable: If two exogeneous variables have very high correlation then the one with the higher correlation with the endogeneous variable will absorb all explanatory ability from other correlated exogeneous variable, like the case of Pepsi and Dow.

**Adjusted** $R^{2}$ address the issue of overfitting/compleity by including one more exogeneous variable in the model. It is always less than the $R^{2}$.

An **Adjusted** $R^{2}$ increases if the degree of the improvement from adding one exogeneous variable is greater than the penalty imposed by introducing the new variable.
