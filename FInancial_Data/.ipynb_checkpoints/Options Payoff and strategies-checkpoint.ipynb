```python
#!pip install option-price
#!pip install opstrat
import math

#optionprice to create option and determine their price
from optionprice import Option

# opstrat another library to develop option strategy
import opstrat as op
```

Arguments for the option class

kind= call/put

type:
eurpoean= True/False

**Numerical**
Stock Price- s0

Strike Level=k

Time to expiration= t

Volatility= sigma (annualised)

Risk-free rate= r (annualised)

Dividend-yield- dv (annualised)

```python
#Creating an ATM European Call

myCall= Option(
    kind="call", european= True, s0=100, k=100, t=30, sigma= 0.20, r= 0.05, dv=0
              )
print(myCall)
```

```python
type(myCall)
```

```python
# Creating an ATM European Put Option

myPut= Option(
    kind="put", european = True, s0=100, k=100, r=0.05, sigma=0.20, t=30, dv=0
)
print(myPut)
```

## Getting Option Price

Methods of getting Option Price (not payoff):

Analytic Method: Employes MAthematics like stochastic calculas, Fourier Transform etc. 

Numeric Method: Numerical/Computation technique ( Binomial Trees, Finite Difference Method, FEM), Randomization(Monte Carlo Sampling, agent- based models)


```python
#Option Price using the Black Scholes Method:
callBS= round(myCall.getPrice(),4)

#Option Price using the Binomial Trees Method:  
callBT= round(myCall.getPrice(method= "BT", iteration= 5000),4)

#Option Price using the Monte Carlo Simulation
callMC= round(myCall.getPrice(method="MC", iteration= 500000),4)

[callBS,callBT,callMC]
```

```python
#Option Price using the Black Scholes Method:
putBS= round(myPut.getPrice(),4)

#Option Price using the Binomial Trees Method:  
putBT= round(myPut.getPrice(method= "BT", iteration= 5000),4)

#Option Price using the Monte Carlo Simulation
putMC= round(myPut.getPrice(method="MC", iteration= 500000),4)

[putBS,putBT,putMC]
```

```python
# Put call parity
# call- put= S-K* e^(-rt)


c= callBS
p=putBS

round(c-p,6)
```

```python
S=100
K=100
r=0.05
t=30/365

round(S-K*math.exp(-r*t),6)
```

**Plotting Option's Payoff:**

Function- single_plotter()

Argumments:
spot: current strike price

strike: strike level

op_type: "c" or "p"

tr_type: (trade side) "b" (buy/long) "s" (selling/short)

op_pr: option price (premium)


```python
op.single_plotter(spot=100, strike=100, op_type="c", tr_type="b", op_pr=callBS)
```

```python
op.single_plotter(spot=100, strike=100, op_type="p", tr_type="s", op_pr=putBS)
```

```python
#Option Combination
# Buying a call and put at the same strike - straddle strategy- trrades volatility
myCall={"op_type": "c", "strike":100, "tr_type": "b", "op_pr": callBS}
myPut={ "op_type": "p", "strike": 100, "tr_type":"b","op_pr": putBS}
op_list=[myCall,myPut]
op.multi_plotter(spot=100,spot_range=10,op_list=op_list)
#spot_range= range of spot variation in %
```

```python
myCall2 = Option(
    european=True, kind="call", s0=100, k=110, t=30, sigma=0.20, r=0.05, dv=0
)
myPut2 = Option(european=True, kind="put", s0=100, k=90, t=30, sigma=0.20, r=0.05, dv=0)
callBS2 = round(myCall2.getPrice(), 4)
putBS2 = round(myPut2.getPrice(), 4)
print([callBS2, putBS2])
```

```python
myCall = {"op_type": "c", "strike": 100, "tr_type": "b", "op_pr": callBS}
myPut = {"op_type": "p", "strike": 100, "tr_type": "b", "op_pr": putBS}
myCall2 = {"op_type": "c", "strike": 110, "tr_type": "s", "op_pr": callBS2}
myPut2 = {"op_type": "p", "strike": 90, "tr_type": "s", "op_pr": putBS2}
op_list = [myCall, myPut, myCall2, myPut2]
op.multi_plotter(spot=100, spot_range=20, op_list=op_list)
```
