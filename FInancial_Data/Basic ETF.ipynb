```python
import datetime

import pandas as pd
import pandas_datareader.data as web
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yfin

yfin.pdr_override()
```

```python
start= datetime.date.today()-datetime.timedelta(365*20)
end=datetime.date.today()

prices= pd.DataFrame(web.DataReader(["XLF"],start,end)["Adj Close"])
prices=prices.rename(columns={"Adj Close":"XLF"})
prices=prices.dropna()
```

```python
plt.figure(figsize=(12,5))

sns.lineplot(x="Date",y="XLF",data=prices,label="Daily XLF 500 prices.")
```

Checking for the loss XLF had during the period of Great Depression from 2006 to start of 2010

```python
drawdown=(
    prices["2006-01-01":"2010-01-01"].max()-prices["2006-01-01":"2010-01-01"].min()
)/prices["2006-01-01":"2010-01-01"].max()
100*drawdown.round(3)
```

XLF lost around 82.75 during this period of great depression

```python
#Overall Return
(100*(prices.XLF[-1]-prices.XLF[0])/prices.XLF[0]).round(2)
```

The overall return in XLF 500 during this period is 141.52%
